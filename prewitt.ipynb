```python
# Step 1: Import libraries
import numpy as np
import matplotlib.pyplot as plt
from skimage import data
from scipy.ndimage import convolve

# Step 2: Load sample coins image
coins = data.coins()

# Step 3: Define Prewitt kernels
prewitt_x = np.array([
    [-1, 0, 1],
    [-1, 0, 1],
    [-1, 0, 1]
])

prewitt_y = np.array([
    [-1, -1, -1],
    [ 0,  0,  0],
    [ 1,  1,  1]
])

# Step 4: Apply convolution
Gx = convolve(coins, prewitt_x)
Gy = convolve(coins, prewitt_y)

# Step 5: Compute magnitude
magnitude = np.sqrt(Gx**2 + Gy**2)

# Step 6: Normalize for display
Gx_disp = np.uint8(np.clip(np.abs(Gx), 0, 255))
Gy_disp = np.uint8(np.clip(np.abs(Gy), 0, 255))
mag_disp = np.uint8(np.clip(magnitude, 0, 255))

# Step 7: Plot results
plt.figure(figsize=(12, 6))

plt.subplot(2,2,1)
plt.title("Original Coins Image")
plt.imshow(coins, cmap='gray')
plt.axis('off')

plt.subplot(2,2,2)
plt.title("Prewitt X (Vertical Edges)")
plt.imshow(Gx_disp, cmap='gray')
plt.axis('off')

plt.subplot(2,2,3)
plt.title("Prewitt Y (Horizontal Edges)")
plt.imshow(Gy_disp, cmap='gray')
plt.axis('off')

plt.subplot(2,2,4)
plt.title("Prewitt Magnitude")
plt.imshow(mag_disp, cmap='gray')
plt.axis('off')

plt.tight_layout()
plt.show()
```
