```python
# Step 1: Import libraries
import cv2
import numpy as np
import matplotlib.pyplot as plt
from skimage import data

# Step 2: Load sample coins image
coins = data.coins()

# Step 3: Apply Sobel filters
sobelx = cv2.Sobel(coins, cv2.CV_64F, 1, 0, ksize=3)
sobely = cv2.Sobel(coins, cv2.CV_64F, 0, 1, ksize=3)

# Step 4: Compute gradient magnitude
sobel_magnitude = np.sqrt(sobelx**2 + sobely**2)

# Step 5: Normalize for display
sobelx_disp = cv2.convertScaleAbs(sobelx)
sobely_disp = cv2.convertScaleAbs(sobely)
sobel_mag_disp = cv2.convertScaleAbs(sobel_magnitude)

# Step 6: Plot results
plt.figure(figsize=(12, 6))

plt.subplot(2,2,1)
plt.title("Original Coins Image")
plt.imshow(coins, cmap='gray')
plt.axis('off')

plt.subplot(2,2,2)
plt.title("Sobel X (Vertical Edges)")
plt.imshow(sobelx_disp, cmap='gray')
plt.axis('off')

plt.subplot(2,2,3)
plt.title("Sobel Y (Horizontal Edges)")
plt.imshow(sobely_disp, cmap='gray')
plt.axis('off')

plt.subplot(2,2,4)
plt.title("Sobel Magnitude")
plt.imshow(sobel_mag_disp, cmap='gray')
plt.axis('off')

plt.tight_layout()
plt.show()
```
