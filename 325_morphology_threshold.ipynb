```python
import numpy as np

import matplotlib.pyplot as plt

from skimage import data

from skimage import img_as_ubyte
from skimage import data, io
from skimage import img_as_ubyte



# 2. LOAD the actual image file into a NumPy array
# (Using as_gray=True if you intend to work with a grayscale phantom image)

image_path = '/content/IMG_9224.JPG'
image = io.imread(image_path, as_gray=True)

# 3. Convert to 8-bit (0–255)
image = img_as_ubyte(image)



# Set threshold

threshold_value = 100

binary_image = np.where(image > threshold_value, 255, 0)



# Create figure

plt.figure(figsize=(15, 4))



# 1️ Original Image

plt.subplot(1, 3, 1)

plt.imshow(image, cmap='gray')

plt.title('Original Phantom Image (0–255)')

plt.axis('off')



# 2️Histogram of Original Image

plt.subplot(1, 3, 2)

plt.hist(image.ravel(), bins=256)

plt.title('Histogram of Original Image')

plt.xlabel('Pixel Intensity')

plt.ylabel('Frequency')

plt.axvline(threshold_value) # Show threshold location

plt.grid()



# 3️⃣ Thresholded Image

plt.subplot(1, 3, 3)

plt.imshow(binary_image, cmap='gray')

plt.title(f'Thresholded Image (T = {threshold_value})')

plt.axis('off')



plt.tight_layout()

plt.show()
```
