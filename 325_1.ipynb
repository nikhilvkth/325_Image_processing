```python
import numpy as np

import matplotlib.pyplot as plt



# Generating a non-periodic signal with three clusters to mimic real-world sensor output

np.random.seed(42) # For reproducibility

time = np.linspace(0, 10, 1000) # Time axis (arbitrary units)



# Creating three clusters of signal activity

signal = np.zeros_like(time)



# First cluster (low voltage, small noise)

signal[:300] = 0.5 + 0.2 * np.random.randn(300)



# Second cluster (higher voltage, more variation)

signal[300:700] = 2.0 + 0.5 * np.random.randn(400)



# Third cluster (medium voltage, moderate noise)

signal[700:] = 1.0 + 0.3 * np.random.randn(300)



# Plot the signal

plt.figure(figsize=(8, 4))

plt.plot(time, signal, color='blue', label="Sensor Output (Voltage)")

plt.xlabel("Time (arbitrary units)")

plt.ylabel("Voltage (V)")

plt.title("Example of an Arbitrary Non-Periodic Electrical Signal from an Image Sensor")

plt.legend()

plt.show()
```

```python
height, width = 50, 1000 # 50 pixels high, 1000 pixels wide



# Normalize the signal to fit in the image intensity range (0 to 255)

signal_normalized = (signal - signal.min()) / (signal.max() - signal.min()) * 255

signal_normalized = signal_normalized.astype(np.uint8)



# Create a 2D image by repeating the 1D signal across the height

image = np.tile(signal_normalized, (height, 1))



# Display the image

plt.figure(figsize=(12, 4))

plt.imshow(image, cmap='gray', aspect='auto')

plt.title("Image Representation of Sensor Signal")

plt.xlabel("Time (arbitrary units)")

plt.ylabel("Sensor Output (Intensity)")

plt.colorbar(label="Intensity (0-255)")

plt.savefig("image_new4.jpg",dpi=300, bbox_inches='tight')

plt.show()
```
