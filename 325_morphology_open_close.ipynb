```python
import numpy as np
import matplotlib.pyplot as plt
from skimage import io, img_as_ubyte
# Import morphology module and a structuring element (kernel)
from skimage import morphology

# 1. Load and threshold the image (from your previous step)
image_path = '/content/IMG_9224.JPG'
image = io.imread(image_path, as_gray=True)
image = img_as_ubyte(image)

threshold_value = 100
binary_image = np.where(image > threshold_value, 255, 0)

# 2. Define a Structuring Element (Kernel)
# We will use a disk-shaped footprint. Increase the radius (e.g., 5 or 10) for stronger effects.
footprint = morphology.disk(radius=5)

# 3. Apply Morphological Operations
# Opening: Erosion followed by Dilation (Removes small bright noise/spots)
opened_image = morphology.opening(binary_image, footprint)

# Closing: Dilation followed by Erosion (Fills small dark holes/gaps)
closed_image = morphology.closing(binary_image, footprint)

# 4. Create the Visualization Figure
plt.figure(figsize=(18, 5))

# Subplot 1: Binary Input
plt.subplot(1, 3, 1)
plt.imshow(binary_image, cmap='gray')
plt.title('Binary Thresholded Image')
plt.axis('off')

# Subplot 2: Morphological Opening
plt.subplot(1, 3, 2)
plt.imshow(opened_image, cmap='gray')
plt.title('After Opening\n(Removes Noise/Isolated Pixels)')
plt.axis('off')

# Subplot 3: Morphological Closing
plt.subplot(1, 3, 3)
plt.imshow(closed_image, cmap='gray')
plt.title('After Closing\n(Fills Holes/Gaps)')
plt.axis('off')

plt.tight_layout()
plt.show()
```
